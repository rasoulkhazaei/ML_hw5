```python
# Define a Python class, KMeansCluster, to implement the K-Means clustering algorithm.
# Include an initialization method that accepts parameters for the number of clusters (k),
# convergence tolerance (tol), and maximum iterations (max_iter).

class KMeansCluster:
    def __init__(self, k, tol, max_iter):
        self.k = k
        self.tol = tol
        self.max_iter = max_iter
        self.centroids = {}
        self.sse = 0
        self.labels = []
        self.data = []
        self.n = 0
        
    def fit(self, data, centroids):
        self.data = data
        self.n = len(data)
        self.centroids = centroids
        
        for i in range(self.max_iter):
            self.labels = []
            for i in range(len(self.data)):
                distances = [np.linalg.norm(self.data[i] - centroid) for centroid in self.centroids]
                cluster = distances.index(min(distances))
                self.labels.append(cluster)
                
            for i in range(self.k):
                points = [self.data[j] for j in range(len(self.data)) if self.labels[j] == i]
                self.centroids[i] = np.mean(points, axis=0)
                
            # sse
            self.sse = 0
            for i in range(len(self.data)):
                self.sse += np.linalg.norm(self.data[i] - self.centroids[self.labels[i]])**2
                
            if self.sse < self.tol:
                break
            
            
    def get_clusters(self):
        return self.labels
    
    def get_centroids(self):
        return self.centroids
    
    def accuracy(self, labels):
        correct = 0
        for i in range(len(labels)):
            if labels[i][0] == self.labels[i]:
                correct += 1
        return correct / len(labels)
    
    def sse(self):
        return self.sse
    
            
        
    
    
# read data from csv file
import pandas as pd
import csv
data = pd.read_csv('Q_4/breast_data.csv')
data = data.values.tolist()

    
# read labels from csv file
lb = pd.read_csv('Q_4/breast_labels.csv')
labels = lb.values.tolist()

# cluster the data with k = 2, tol = 0.001, max_iter = 100
k = 2
import numpy as np
tol = 0.001
max_iter = 100
kmeans = KMeansCluster(k, tol, max_iter)
kmeans.fit(data, np.ones((k, len(data[0]))))
clusters = kmeans.get_clusters()
print('Clusters:', clusters)
print('Accuracy:', kmeans.accuracy(labels))
print('SSE:', kmeans.sse(data))
```
